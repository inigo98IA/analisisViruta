Eliminamos las imagenesque no tengan un foemato válido

```
for c in ['hierro','cobre','madera']:
  print(c)
  verify_images(path/c,delete=True,max_size=700)
```

Particionamos el dataset  y usamos un 20% como dataset de validación (parámetro valid_pct), se aplican también distintas técnicas de aumento (parámetro ds_tfms, crea una nueva celda y ejecuta la instrucción get_transforms() para ver las técnicas de aumento que se están utilizando), las imágenes se reescalan a tamaño 224 (este es el tamaño de entrada que espera que tengan las imágenes la red neuronal que construiremos un poco más adelante), y se normalizan utilizando el dataset de ImageNet.

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path,train=".",valid_pct=0.2,ds_tfms=get_transforms(),size=224,num_workers=4).normalize(imagenet_stats)
```

Mostramos un batch del dataset que acabamos de cargar.

```
data.show_batch(rows=3,figsize=(7,8))
```

Creamos nuestra red convolucional usando la arquitectura ResNet 34

```
learn = create_cnn(data,models.resnet34,metrics=[error_rate,accuracy])
```

Actualizamos los pesos de la ultimas capas

```
learn.fit_one_cycle(4)
```

Guardamos el estado de nuestra red

```
learn.save('stage-1')
```

Descongelamos las capas de nuestra red

```
learn.unfreeze()
```

Vamos a actualizar las capas del principio con un learn-rate pequeño y las últimas con uno más grande.

```
learn.fit_one_cycle(4,max_lr=slice(3e-5,3e-4))
```

Lo guardamos y lo cargamos

```
learn.save('stage-2')
```

```
learn.load('stage-2')
```

Creamos una interpretación de los resultados obtenidos

```
interp = ClassificationInterpretation.from_learner(learn)
```

Mostramos la matriz de confusión obtenida

```
interp.plot_confusion_matrix()
```
